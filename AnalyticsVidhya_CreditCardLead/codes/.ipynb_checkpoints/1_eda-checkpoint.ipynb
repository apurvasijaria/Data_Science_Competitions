```python
pd.pivot_table(train, index = ['Vintage_yrs'],columns = ['Is_Lead','age_group'],values = ['ID'], aggfunc = 'count')
## Younger people have low agee - early investors
```

```python
pd.pivot_table(train, index = ['Vintage_yrs'],columns = ['Is_Lead'],values = ['ID'], aggfunc = 'count')
## Younger people have low agee - early investors
```

```python
pd.pivot_table(train, index = ['Vintage_yrs'],columns = ['age_group'],values = ['ID'], aggfunc = 'count')
## Younger people have low agee - early investors
```

```python
pd.pivot_table(train, index = ['Gender'],columns = ['age_group'],values = ['ID'], aggfunc = 'count')
## more young males and older females 
```

```python
pd.pivot_table(train, index = ['Gender'],columns = ['Is_Lead'],values = ['ID'], aggfunc = 'count')
## more young females and older males 
```

```python
pd.pivot_table(train, index = ['Is_Active'],columns = ['Is_Lead'],values = ['ID'], aggfunc = 'count')
```

```python
pd.pivot_table(train, index = ['Credit_Product'],columns = ['Is_Lead'],values = ['ID'], aggfunc = 'count')
```

```python
pd.pivot_table(train, index = ['Credit_Product'],columns = ['Is_Active'],values = ['ID'], aggfunc = 'count')
```

```python
pd.pivot_table(train, index = ['Vintage_yrs'],columns = ['age_group'],values = ['Avg_Account_Balance'], aggfunc = 'mean')
```
